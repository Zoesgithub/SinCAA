```python
import py3Dmol

def show_pdb(pdb_file):
    with open(pdb_file, "r") as f:
        pdb_data = f.read()
    
    viewer = py3Dmol.view(width=600, height=400)
    viewer.addModel(pdb_data, "pdb")  
    viewer.setStyle({"cartoon": {"color": "spectrum"}})  # 使用 Cartoon 风格
    viewer.zoomTo()
    viewer.show()

# 示例：
show_pdb("tmp.pdb")
```
